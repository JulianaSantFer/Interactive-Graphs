```python
data = pd.read_csv("data.csv")
new_data = data.copy()
data.head(3)
```

```python
overall = data.Overall.value_counts(sort=False).reset_index()
overall.sort_values('index', inplace = True)

x = overall.iloc[:,0]
y = overall.iloc[:,1]

x = x.tolist()
y = y.tolist()
```

```python
trace1 = go.Bar(x = x, y = y, name = 'Overall', opacity = 0.7, marker=dict(color='rgb(178,34,34)'))
data_a = [trace1]

layout = go.Layout(barmode = "group")
fig1 = go.Figure(data = data_a, layout = layout)
```

```python
iplot(fig1)
plotly.offline.plot(fig1, filename='OverallPerformance.html')
```

```python
new_index=(new_data["Overall"].sort_values(ascending=False)).index.values
sortedData=new_data.reindex(new_index)
best_players=sortedData.head(100)

trace1=go.Scatter(
                    x=best_players.Name,
                    y=best_players.ShortPassing,
                    mode="lines",
                    name="Crossing",
                    marker=dict(color="rgba(25,125,25,0.5)"),
                    text=best_players.Nationality)
trace2=go.Scatter(
                    x=best_players.Name,
                    y=best_players.Crossing,
                    mode="lines",
                    name="Dribbling",
                    marker=dict(color="rgba(109,99,109,0.5)"),
                    text=best_players.Nationality)
data=[trace1,trace2]
layout=dict(title="Comparision between Crossing and Dribbling" ,xaxis=dict(ticklen=5,zeroline=False) )

fig=dict(data=data,layout=layout)

iplot(fig)
plotly.offline.plot(fig, filename='CrossingandDribbling.html')
```
